```python
# 1번
import pandas as pd
# DataFrame() 함수로 데이터프레임 변환. 변수 df에 저장
exam_data = {'이름' : [ '서준', '우현', '인아'],
             '수학' : [ 90, 80, 70],
             '영어' : [ 98, 89, 95],
             '음악' : [ 85, 95, 100],
             '체육' : [ 100, 90, 90]}
org_df = pd.DataFrame(exam_data)
# '이름' 열을 새로운 인덱스로 지정하고, df 객체에 변경사항 반영
org_df = org_df.set_index('이름')
print(org_df)
```

```python
# 1번
import pandas as pd
import matplotlib.pyplot as plt

# 파일 읽기
file_path = "./data/남북한발전전력량.xlsx"
df = pd.read_excel(file_path)

# 남한 데이터 필터링 및 총 전력량 계산
south_korea = df[df['국가'] == '남한']
total_power_south = south_korea.groupby('연도')['전력량'].sum()

# 선 그래프 그리기
plt.figure(figsize=(10, 6))
total_power_south.plot(kind='line', marker='o')
plt.title('남한의 연도별 총 전력량 변화')
plt.xlabel('연도')
plt.ylabel('총 전력량 (GWh)')
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 파일 읽기
file_path = "./data/남북한발전전력량.xlsx"
df = pd.read_excel(file_path)

# 데이터프레임의 열 확인
print(df.columns)

# 남한 데이터 필터링 및 총 전력량 계산 (열 이름을 확인한 후 수정)
south_korea = df[df['국가'] == '남한']  # '국가'가 올바른 열 이름인지 확인
total_power_south = south_korea.groupby('연도')['전력량'].sum()

# 선 그래프 그리기
plt.figure(figsize=(10, 6))
total_power_south.plot(kind='line', marker='o')
plt.title('남한의 연도별 총 전력량 변화')
plt.xlabel('연도')
plt.ylabel('총 전력량 (GWh)')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc

# 한글 폰트 설정
rc('font', family='AppleGothic')  # Mac에서 사용할 경우

plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 깨짐 방지


# 1.문제: 남한의 연도별 총전력량(합계) 변화를 선그래프 그리기
df_ns = df.iloc[[0], 3:] 

# 행 인덱스 변경
df_ns.index = ['남한발전량'] 
# print(df_ns.head())
df_graph = df_ns.T
df_graph.plot( title="남한 총발전량")
```

```python
#2. 문제: 북한의 연도별 총 전력량(합계)을 막대 그래프로 그려라.
df_ns = df.iloc[[5], 3:] 

# 행 인덱스 변경
df_ns.index = ['북한발전량'] 
# print(df_ns.head())
df_graph = df_ns.T
df_graph.plot(kind='bar', title="북한 총발전량")
```

```python
#3. 문제: 남한의 연도별 수력발전량 변화를 꺽은선 그래프로 그려라.
df_ns = df.iloc[[1], 3:] 

# 행 인덱스 변경
df_ns.index = ['남한 수력발전량'] 
# print(df_ns.head())
df_graph = df_ns.T
df_graph.plot(kind='line', title="남한 수수력 발전량")
```

```python
#4. 문제: 북한의 연도별 화력발전량 변화를 꺽은선 그래프로 그려라.
df_ns = df.iloc[[7], 3:] 

# 행 인덱스 변경
df_ns.index = ['북한 화력발전량'] 
# print(df_ns.head())
df_graph = df_ns.T
df_graph.plot(kind='bar', title="북한 화력 발전량")
```

```python
#5. 문제: 남한의 화력과 원자력 발전량의 연도별 비교를 선그래프로 그려라.
df_ns = df.iloc[[2,3], 3:] 

# 행 인덱스 변경
df_ns.index = ['화력발전량','원자력'] 
# print(df_ns.head())
df_graph = df_ns.T
ax = df_graph.plot(kind='line' , title="남한 화력과 원자력발전량 년도별 비교")
ax.legend(title='발전종류')
```
