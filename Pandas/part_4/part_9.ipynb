```python
import pandas as pd
import matplotlib.pyplot as plt

# 날짜 범위 생성
data_range = pd.date_range(start='2024-10-01', periods=10, freq='D')
data = [1, 2, 5, 8, 4, 6, 4, 8, 7, 8]

# DataFrame 생성
df = pd.DataFrame(data, index=data_range, columns=["value"])  # 리스트 형태로 수정

# 막대 그래프 그리기
df.plot(kind='line', title='Sample Chart')
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation=45)  # x축 레이블 각도 조정
plt.tight_layout()  # 레이아웃 자동 조정
plt.show()
```
